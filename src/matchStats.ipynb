```python
palyerWeiHei = pd.DataFrame(matchLineups_data, columns = ['match_id', 'team', 'weight', 'height'])

# df.loc[df.Weight == "155", "Name"] = "John"

# rslt_df = dataframe[dataframe['Percentage'] > 80] 
print(palyerWeiHei.size)
palyerWeiHei = palyerWeiHei.loc[palyerWeiHei.weight < 120]
palyerWeiHei = palyerWeiHei.loc[palyerWeiHei.weight > 50]
palyerWeiHei = palyerWeiHei.loc[palyerWeiHei.height > 140]
print(palyerWeiHei.size)

plt.title("Player weight & height") 
plt.xlabel("weight") 
plt.ylabel("height")
plt.plot(palyerWeiHei['weight'], palyerWeiHei['height'], "ob")
plt.show()
```

```python
mean_palyerWeiHei = palyerWeiHei.groupby(['match_id', 'team'] , as_index = False).mean()
print(mean_palyerWeiHei.size)

mean_palyerWeiHei_array = {}

home = mean_palyerWeiHei.loc[mean_palyerWeiHei.team == 'home']
away = mean_palyerWeiHei.loc[mean_palyerWeiHei.team == 'away']

for index, row in home.iterrows():
    mean_palyerWeiHei_array[row['match_id']] = [round(row['weight'], 2) , round(row['height'], 2), 0, 0]
for index, row in away.iterrows():
    if row['match_id'] in mean_palyerWeiHei_array:
        mean_palyerWeiHei_array[row['match_id']][2] = round(row['weight'], 2)
        mean_palyerWeiHei_array[row['match_id']][3] = round(row['height'], 2)

print(home.size)
print(away.size)

data_file = open(os.getcwd()[:-4] + '\\cleanedData\\meanPalyerWeiHei.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(data_file)

rowData = {
  'match_id': [], 'home_player_avg_wei': [], 'home_player_avg_hei': [], 'away_player_avg_wei': [], 'away_player_avg_hei': []
}

# insert header
csv_writer.writerow(['match_id', 'home_player_avg_wei', 'home_player_avg_hei', 'away_player_avg_wei', 'away_player_avg_hei'])

# insert data
for key, values in mean_palyerWeiHei_array.items():
#     print(key, '->', values)
    row = [key]
    for value in values:
        row.append(value)
    csv_writer.writerow(row)

        
print('done mean_palyerWeiHei')
```

```python
records = []
for i in range(0, len(df.index)):
#     print(i)
    records.append([str(df.values[i,j]) for j in range(0, 6)])
```

```python
records
records_df = pd.DataFrame(records)
records_df
```

```python
# min_support = 0.00625
# an item frequency a least appear 23 time in 3825 = 0.00625
min_support = 0.00625
# likelihood that an item B is also bought if item A appear together = 30%
min_confidence = 0.3
# ratio of item A and item B appear in cart together
min_lift = 3
# a least items in the rules
min_length = 4

association_rules = apriori(records, min_support=min_support, min_confidence=min_confidence, min_lift=min_lift, min_length=min_length)
association_results = list(association_rules)
```

```python
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    
    if len(items) >= min_length:
        printItems = ""
        for key in items:
            printItems +=  key + " -> "

        printItems = printItems[:-3]
        print("Rule: " + printItems )

        print("Support: " + str(item[1]))

        print("Confidence: " + str(item[2][0][2]))
        print("Lift: " + str(item[2][0][3]))
        print("=====================================")
```

```python
# 導入函式庫
import os
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.datasets import mnist
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.utils import np_utils  # 用來後續將 label 標籤轉為 one-hot-encoding  
from matplotlib import pyplot as plt
```

```python
# target data
pre_lea_2021_data = pd.read_csv(os.getcwd() + '\\pre_lea_2021_data.csv')
pre_lea_2021_data
```

```python
X = pre_lea_2021_data.drop(['result'], axis = 1).copy()

X_encoded = pd.get_dummies(X, columns=['HF', 'AF'])
X_encoded
```

```python
X_encoded.values
```

```python
y = pre_lea_2021_data['result'].copy()

# def one_hot_y_label(y):
#     result = np.zeros((len(y), 3), dtype=int)
#     for ind, rs in y.items():
#         if rs == 'home':
#             result[ind][0] = 1
#         elif rs == 'away':
#             result[ind][2] = 1
#         else:
#             result[ind][1] = 1
            
#     return result.astype('float32')

def one_hot_y_label(y):
    result = np.zeros(len(y), dtype=int)
    for ind, rs in y.items():
        if rs == 'home':
            result[ind] = 1
#         else:
#             result.append(0)
    return result
#     return result.astype('float32')


# 將 training 的 label 進行 one-hot encoding，例如數字 7 經過 One-hot encoding 轉換後是 0000001000，即第7個值為 1
y_encode = one_hot_y_label(y)
y_encode
```

```python
# define the keras model
model = Sequential()
model.add(Dense(312, input_dim=936, activation='relu'))
model.add(Dense(104, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X_encoded.values, y_encode, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_encoded.values, y_encode)
print('Accuracy: %.2f' % (accuracy*100))
```

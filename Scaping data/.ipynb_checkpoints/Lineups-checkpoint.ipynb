```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import requests
from bs4 import BeautifulSoup
```

```python
# Starting with latest data for Spanish league, because I'm a Barcelona fan
# url = base_url+'/'+leagues[0]+'/'+seasons[1]

url = 'https://fbref.com/en/comps/9/schedule/Premier-League-Scores-and-Fixtures#sched_ks_10728_1'
# url = 'https://www.football-lineups.com/match/368165/'

res = requests.get(url)
soup = BeautifulSoup(res.content, "lxml")
```

```python
# Based on the structure of the webpage, I found that data is in the JSON variable, under <script> tags
trs = soup.find_all('tr')

ass = []

# Check our <script> tags
for el in trs:
    print('*'*50)
    print(el.find(class_="center").find("a"))
    ass.append(el.find(class_="center").find("a"))
```

```python
columns = ["href"]
ass[1]['href']

hrefs = []
for a_tag in ass:
    if a_tag != None:
        hrefs.append("https://fbref.com" + a_tag['href'])

hrefs
```

```python
len(hrefs)
```

```python
fulham_arsenal_match = hrefs[0]

res = requests.get(fulham_arsenal_match)
match_soup = BeautifulSoup(res.content, "lxml")

tables = match_soup.find_all(class_="table_wrapper")

home_trs = tables[0].find("tbody").find_all("tr")
away_trs = tables[7].find("tbody").find_all("tr")
# for el in tables:
#     print('*'*50)
#     print(el.text)
away_trs
```

```python
h_playerName = []
h_playerNum = []
h_playerPos = []
for tr in home_trs:
    th = tr.find("th")
    tds = tr.find_all("td")
    
    h_playerName.append(th.text)
    h_playerNum.append(tds[0].text)
    h_playerPos.append(tds[2].text)
        
a_playerName = []
a_playerNum = []
a_playerPos = []
for tr in away_trs:
    th = tr.find("th")
    tds = tr.find_all("td")
    
    a_playerName.append(th.text)
    a_playerNum.append(tds[0].text)
    a_playerPos.append(tds[2].text)
        
a_playerName
```

```python
columns = ['name', 'num', 'pos']

data = pd.DataFrame(columns = columns)
for index, name in enumerate(h_playerName):
    if "\xa0\xa0\xa0" in name:
        continue
    number = h_playerNum[index]
    pos = h_playerPos[index]
    
    player_data = pd.DataFrame([[name, number, pos]])
    player_data.columns = columns
    data = data.append(player_data, ignore_index=True)
    
df = pd.DataFrame(data, columns=columns)
# df.head(2)
df
```

```python
columns = [
    "HF",
    "H1N", 'H1P', "H2N", 'H2P', "H3N", 'H3P', "H4N", 'H4P', "H5N", 'H5P', "H6N", 'H6P', 
    "H7N", 'H7P', "H8N", 'H8P', "H9N", 'H9P', "H10N", 'H10P', "H11N", 'H11P',

    "AF",
    "F1N", 'F1P', "F2N", 'F2P', "F3N", 'F3P', "F4N", 'F4P', "F5N", 'F5P', "F6N", 'F6P', 
    "F7N", 'F7P', "F8N", 'F8P', "F9N", 'F9P', "F10N", 'F10P', "F11N", 'F11P',
    
    "HS", "AS", "result"
]

data = pd.DataFrame(columns = columns)

pType = []

count = 0
for href in hrefs:
    match = href

    res = requests.get(match)
    match_soup = BeautifulSoup(res.content, "lxml")

    tables = match_soup.find_all(class_="table_wrapper")

    home_trs = tables[0].find("tbody").find_all("tr")
    away_trs = tables[7].find("tbody").find_all("tr")
    
    h_playerName = []
    h_playerNum = []
    h_playerPos = []
    for tr in home_trs:
        th = tr.find("th")
        tds = tr.find_all("td")

        h_playerName.append(th.text)
        h_playerNum.append(tds[0].text)
        h_playerPos.append(tds[2].text)
            
        if tds[2].text not in pType:
            pType.append(tds[2].text)

    a_playerName = []
    a_playerNum = []
    a_playerPos = []
    for tr in away_trs:
        th = tr.find("th")
        tds = tr.find_all("td")

        a_playerName.append(th.text)
        a_playerNum.append(tds[0].text)
        a_playerPos.append(tds[2].text)
        
    lineups = match_soup.find_all(class_="lineup")

    home_formation = lineups[0].find_all("tr")[0].text
    home_formation = home_formation.split("(")
    home_formation = home_formation[1]
    home_formation = home_formation[:-1]

    away_formation = lineups[1].find_all("tr")[0].text
    away_formation = away_formation.split("(")
    away_formation = away_formation[1]
    away_formation = away_formation[:-1]
    
    

    match_lineup = []
    
    match_lineup.append(home_formation)
    for index, name in enumerate(h_playerName):
        if "\xa0\xa0\xa0" in name:
            continue
        number = h_playerNum[index]
        pos = h_playerPos[index]

        match_lineup.append(number)
        if ',' in pos:
            p = pos.split(",")
            match_lineup.append(p[0])
        else:
            match_lineup.append(pos)

    match_lineup.append(away_formation)
    for index, name in enumerate(a_playerName):
        if "\xa0\xa0\xa0" in name:
            continue
        number = a_playerNum[index]
        pos = a_playerPos[index]

        match_lineup.append(number)
        if ',' in pos:
            p = pos.split(",")
            match_lineup.append(p[0])
        else:
            match_lineup.append(pos)
    
    scores = match_soup.find_all(class_="score")

    result = "None"
    if scores[0].text > scores[1].text:
        result = "home"
    elif scores[0].text < scores[1].text:
        result = "away"
    else:
        result = "draw"
    
    match_lineup.append(scores[0].text)
    match_lineup.append(scores[1].text)
    match_lineup.append(result)

    players_data =  pd.DataFrame([match_lineup])
    players_data.columns = columns
    data = data.append(players_data, ignore_index=True)
    
    count = count + 1
    print("Loaded so far: "+str(count)+"/"+str(len(hrefs)), end="\r")
    if count == 213:
        break

data.to_csv('pre_lea_2021_lineUps.csv', encoding='utf-8-sig', index=False)

data
```

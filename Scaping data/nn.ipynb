```python
# 導入函式庫
import os
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.datasets import mnist
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.utils import np_utils  # 用來後續將 label 標籤轉為 one-hot-encoding  
from matplotlib import pyplot as plt
```

```python
# target data
pre_lea_2021_data = pd.read_csv(os.getcwd() + '\\pre_lea_2021_data.csv')
pre_lea_2021_data
```

```python
X = pre_lea_2021_data.drop(['result'], axis = 1).copy()

X_encoded = pd.get_dummies(X, columns=['HF', 'AF'])
X_encoded
```

```python
y = pre_lea_2021_data['result'].copy()

# def one_hot_y_label(y):
#     result = np.zeros((len(y), 3), dtype=int)
#     for ind, rs in y.items():
#         if rs == 'home':
#             result[ind][0] = 1
#         elif rs == 'away':
#             result[ind][2] = 1
#         else:
#             result[ind][1] = 1
            
#     return result.astype('float32')

def one_hot_y_label(y):
    result = np.zeros(len(y), dtype=int)
    for ind, rs in y.items():
        if rs == 'home':
            result[ind] = 1
#         elif rs == 'away':
#             result[ind] = 1
#         else:
#             result[ind] = 1
            
    return result.astype('float32')


# 將 training 的 label 進行 one-hot encoding，例如數字 7 經過 One-hot encoding 轉換後是 0000001000，即第7個值為 1
y_encode = one_hot_y_label(y)
y_encode
```

```python
# 建立簡單的線性執行的模型
# model = Sequential()
# # Add Input layer, 隱藏層(hidden layer) 有 256個輸出變數
# # model.add(Dense(units=256, input_dim=28*28, kernel_initializer='normal', activation='relu')) 
# model.add(Dense(units=300, input_dim=950, kernel_initializer='normal', activation='relu')) 
# # Add output layer
# model.add(Dense(units=3, kernel_initializer='normal', activation='softmax'))

model = Sequential()
model.add(Dense(600, input_dim=950, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 編譯: 選擇損失函數、優化方法及成效衡量方式
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
def poundsToMetric(pounds):
    kilograms = pounds / 2.2
    return int(kilograms)

def feetToCentimeters(h_ft, h_inch):
    h_inch += h_ft * 12
    return math.ceil(h_inch * 2.54)

def find_between(s, first, last):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""

def dateConvert(date):
    datetime_object = datetime.strptime(date, '%b %d, %Y')
    return datetime_object.strftime("%Y") + '-' + datetime_object.strftime("%m") + '-' + datetime_object.strftime("%d")

def priceFormation(price):
    if not price.isnumeric():
        lasts = price[-1:]
        if lasts == "M":
            #add 000 000
            p = price[:-1].split('.')
            addZero = 6 - len(p[1]) if len(p) == 2 else 6
        elif lasts == "K":
            #add 000 
            p = price[:-1].split('.')
            addZero = 3 - len(p[1]) if len(p) == 2 else 3
        else:
            print('price error' , price)
        
        zeros = ''.join('0' for i in range(addZero))
        reP = p[0] + p[1] + zeros if len(p) == 2 else p[0] + zeros
        
        return int(reP)
    else:
        return int(price)
    
```
